```python
# Identificar y mostrar los outliers en cada columna
def calcular_outliers(df, columna):
    if df[columna].dtype == 'object':
        return None, None, None, None
    else:
        Q1 = df[columna].quantile(0.25)
        Q3 = df[columna].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df[(df[columna] < Q1 - 1.5 * IQR) | (df[columna] > Q3 + 1.5 * IQR)]
        return outliers, Q1, Q3, IQR

# Calcular outliers para todas las columnas numÃ©ricas
for col in df.columns:
    outliers, Q1, Q3, IQR = calcular_outliers(df, col)
    if outliers is not None:
        print(f"Outliers en {col}: {len(outliers)}")
    else:
        print(f"Columna '{col}' es de tipo object, saltando el cÃ¡lculo de outliers.")
```

```python
# Calcular los cuartiles
Q1 = df['id'].quantile(0.25)
Q3 = df['id'].quantile(0.75)
IQR = Q3 - Q1

# Calcular los lÃ­mites para identificar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los outliers
outliers = df[(df['id'] < lower_bound) | (df['id'] > upper_bound)]

# Mostrar los outliers
print("Outliers:")
outliers
```

# PREPROCESAMIENTO

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
# Separar caracterÃ­sticas (X) y variable objetivo (y)
X = df.drop(['diagnosis'], axis=1)
y = df['diagnosis']
```

```python
# NormalizaciÃ³n de las caracterÃ­sticas
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Dividir los datos en conjunto de entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# MODELO

```python
# Definir el modelo MLP
model = Sequential()
```

```python
# Agregar la capa de entrada con 32 caracteristicas
model.add(Dense(32, input_dim=X_train.shape[1], activation='relu')) #capa oculta con 32 neuronas

# Agregar la capa de salida con funcion de activacion sigmoide
model.add(Dense(1, activation='sigmoid')) # solo una neurona para clasificacion binaria
```

```python
# Resumen de la red neuronal creada
model.summary()
```

```python
# Compilar el modelo usando la funciÃ³n de pÃ©rdida binary_crossentropy y el optimizador Adam
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
# Entrenar el modelo durante 50 Ã©pocas
history = model.fit(X_train, y_train, epochs=50, validation_split=0.2, batch_size=10, verbose=1)
```

```python
plt.figure(figsize=(12, 4))

# PÃ©rdida
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='PÃ©rdida de Entrenamiento')
plt.plot(history.history['val_loss'], label='PÃ©rdida de ValidaciÃ³n')
plt.xlabel('Ã‰pocas')
plt.ylabel('PÃ©rdida')
plt.title('Curva de PÃ©rdida')
plt.legend()

plt.show()
```

COMENTARIO

Curva de PÃ©rdida:

La pÃ©rdida disminuye tanto en el conjunto de entrenamiento como en el de validaciÃ³n a medida que avanzan las Ã©pocas, indicando que el modelo estÃ¡ aprendiendo correctamente.

La curva de pÃ©rdida de validaciÃ³n se estabiliza alrededor de 0.1, lo que sugiere que el modelo no estÃ¡ sobreajustando (overfitting).

```python
plt.figure(figsize=(12, 4))

# PrecisiÃ³n
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='PrecisiÃ³n de Entrenamiento')
plt.plot(history.history['val_accuracy'], label='PrecisiÃ³n de ValidaciÃ³n')
plt.xlabel('Ã‰pocas')
plt.ylabel('PrecisiÃ³n')
plt.title('Curva de PrecisiÃ³n')
plt.legend()

plt.show()
```

COMENTARIO

Curva de PrecisiÃ³n:

La precisiÃ³n aumenta rÃ¡pidamente durante las primeras Ã©pocas y se estabiliza cerca del 100% en el conjunto de entrenamiento.

La precisiÃ³n de validaciÃ³n es consistente con el entrenamiento y se mantiene por encima de 95%, indicando que el modelo generaliza bien en el conjunto de prueba.

```python
# Evaluar el modelo en el conjunto de prueba
y_pred = (model.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Definir etiquetas para que aparezcan como "Benigno" y "Maligno"
target_names = ['Benigno', 'Maligno']

cm=confusion_matrix(y_test,y_pred)
accuracy = accuracy_score(y_test, y_pred)
print(f'Precision del Modelo: {accuracy:.4f}')
print("Classification report for MLP")
print(classification_report(y_test, y_pred, target_names=target_names))

cm2=ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax=plt.subplots(figsize=(8,6))
cm2.plot(ax=ax)
plt.title('Matriz de confusion Tumor Benigno/Maligno')
plt.ylabel('Valor Real')
plt.xlabel('Valor Predicho')
plt.show()
```
