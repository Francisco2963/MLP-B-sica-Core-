```python
# Identificar y mostrar los outliers en cada columna
def calcular_outliers(df, columna):
    if df[columna].dtype == 'object':
        return None, None, None, None
    else:
        Q1 = df[columna].quantile(0.25)
        Q3 = df[columna].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df[(df[columna] < Q1 - 1.5 * IQR) | (df[columna] > Q3 + 1.5 * IQR)]
        return outliers, Q1, Q3, IQR

# Calcular outliers para todas las columnas numéricas
for col in df.columns:
    outliers, Q1, Q3, IQR = calcular_outliers(df, col)
    if outliers is not None:
        print(f"Outliers en {col}: {len(outliers)}")
    else:
        print(f"Columna '{col}' es de tipo object, saltando el cálculo de outliers.")
```

```python
# Calcular los cuartiles
Q1 = df['id'].quantile(0.25)
Q3 = df['id'].quantile(0.75)
IQR = Q3 - Q1

# Calcular los límites para identificar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los outliers
outliers = df[(df['id'] < lower_bound) | (df['id'] > upper_bound)]

# Mostrar los outliers
print("Outliers:")
outliers
```

# PREPROCESAMIENTO

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
# Separar características (X) y variable objetivo (y)
X = df.drop(['diagnosis'], axis=1)
y = df['diagnosis']
```

```python
# Normalización de las características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Dividir los datos en conjunto de entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# MODELO

```python
# Definir el modelo MLP
model = Sequential()
```

```python
# Agregar la capa de entrada con 32 caracteristicas
model.add(Dense(32, input_dim=X_train.shape[1], activation='relu')) #capa oculta con 32 neuronas

# Agregar la capa de salida con funcion de activacion sigmoide
model.add(Dense(1, activation='sigmoid')) # solo una neurona para clasificacion binaria
```

```python
# Resumen de la red neuronal creada
model.summary()
```

```python
# Compilar el modelo usando la función de pérdida binary_crossentropy y el optimizador Adam
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
# Entrenar el modelo durante 50 épocas
history = model.fit(X_train, y_train, epochs=50, validation_split=0.2, batch_size=10, verbose=1)
```

```python
plt.figure(figsize=(12, 4))

# Pérdida
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Pérdida de Entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida de Validación')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.title('Curva de Pérdida')
plt.legend()

plt.show()
```

COMENTARIO

Curva de Pérdida:

La pérdida disminuye tanto en el conjunto de entrenamiento como en el de validación a medida que avanzan las épocas, indicando que el modelo está aprendiendo correctamente.

La curva de pérdida de validación se estabiliza alrededor de 0.1, lo que sugiere que el modelo no está sobreajustando (overfitting).

```python
plt.figure(figsize=(12, 4))

# Precisión
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Precisión de Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisión de Validación')
plt.xlabel('Épocas')
plt.ylabel('Precisión')
plt.title('Curva de Precisión')
plt.legend()

plt.show()
```

COMENTARIO

Curva de Precisión:

La precisión aumenta rápidamente durante las primeras épocas y se estabiliza cerca del 100% en el conjunto de entrenamiento.

La precisión de validación es consistente con el entrenamiento y se mantiene por encima de 95%, indicando que el modelo generaliza bien en el conjunto de prueba.

```python
# Evaluar el modelo en el conjunto de prueba
y_pred = (model.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Definir etiquetas para que aparezcan como "Benigno" y "Maligno"
target_names = ['Benigno', 'Maligno']

cm=confusion_matrix(y_test,y_pred)
accuracy = accuracy_score(y_test, y_pred)
print(f'Precision del Modelo: {accuracy:.4f}')
print("Classification report for MLP")
print(classification_report(y_test, y_pred, target_names=target_names))

cm2=ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax=plt.subplots(figsize=(8,6))
cm2.plot(ax=ax)
plt.title('Matriz de confusion Tumor Benigno/Maligno')
plt.ylabel('Valor Real')
plt.xlabel('Valor Predicho')
plt.show()
```

El valor de accuracy es 0.9737, lo que indica que el modelo es capaz de clasificar correctamente el 97.37% de los casos en el conjunto de prueba. Este valor se calcula como la proporción de predicciones correctas respecto al total de observaciones. Un valor cercano a 1 sugiere que el modelo tiene un alto rendimiento.
